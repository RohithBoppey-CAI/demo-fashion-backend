```python
import pandas as pd
```

```python
# read from the CSV -> make a dataframe out of it

df = pd.read_csv('./data/fashion_products.csv')
print(df)

# gives 5 entries on default 
print(df.head())
```

```python
# this gives information about the dataframe -> something like describe for a sql table

print(df.info())
```

```python
no_nulls = df.dropna()

# This gives you -> in place result
# df.dropna(inplace=True)

# df.shape gives you number of rows and cols
print(df.shape)
print(no_nulls.shape)   

# print nulls in each

# print(df['price'])
# print(df['price'].isnull())
print(df['price'].isnull().sum())
```

```python
print(df.category.unique())

print(df.category.nunique())

len(df.category.unique())
```

```python
# We want to replace the category wise null values with -> median of the respective column

unique_cat = df.category.unique()
# print(unique_cat)

category_values = df[df['category']=='Women']
# print(category_values.price.median())

replaced = df[df['category']=='Women'].price.fillna(category_values.price.median())
# print(replaced)

for cat in unique_cat:
    cat_entries = df[df['category']==cat]
    cat_entries.price = cat_entries.price.fillna(cat_entries.price.median())
    cat_entries.quantity_sold = cat_entries.quantity_sold.fillna(cat_entries.quantity_sold.median())
    cat_entries.rating = cat_entries.rating.fillna(cat_entries.rating.mean())

    df[df['category']==cat] = cat_entries

print(df)


```

```python
# save the CSV to use it in te fast api application
df.to_csv('./data/processed.csv')
```
